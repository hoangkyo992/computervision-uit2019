```python
import numpy as np
import urllib
import matplotlib.pyplot as plt
import cv2
import sys
import os.path

FACE_CLASSIFIER = 'my_haarcascade_frontalface_default.xml'

def face_detect(img, name):
    source = cv2.imread(img)
    cvt = cv2.cvtColor(source, cv2.COLOR_BGR2GRAY)

    face_cascade = cv2.CascadeClassifier(FACE_CLASSIFIER)
    faces = face_cascade.detectMultiScale(cvt, 1.25, 3)
    cropped_images = []
    idx = 0
    for (x, y, w, h) in faces:
        cropped_img = source[y:y+h, x:x+w]
        cropped_images.append(cropped_img)
        cv2.imwrite(TARGET_DIR + name + '_face_' + str(idx) + '.jpg', cropped_img);
        idx += 1

    for (x, y, w, h) in faces:
        cv2.rectangle(source, (x, y), (x + w, y + h), (255, 255, 0), 4)


    plt.axis('off')
    plt.title('Source image')
    plt.imshow(cv2.cvtColor(source, cv2.COLOR_BGR2RGB))
    plt.show()

    plt.title('Faces detected')
    idx = 0
    for (cropped_img) in cropped_images:
        plt.title(name + '_face_' + str(idx))
        plt.imshow(cv2.cvtColor(cropped_img, cv2.COLOR_BGR2RGB))
        plt.show()
        idx += 1
        
TARGET_DIR = "FaceDB2/Faces/"
IMAGE_DIR = "FaceDB2/"

for filename in os.listdir(IMAGE_DIR):
    if filename.lower().endswith(".jpg"):
        name = filename[0:len(filename) - 4]
        face_detect(IMAGE_DIR + filename, name)
```
