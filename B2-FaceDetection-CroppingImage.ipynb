```python
import numpy as np
import urllib
import matplotlib.pyplot as plt
import cv2
import sys
import os.path
from urllib.request import urlopen

source = cv2.imread("people.jpg")

cvt = cv2.cvtColor(source, cv2.COLOR_BGR2GRAY)
face_classifier = 'my_haarcascade_frontalface_default.xml'

face_cascade = cv2.CascadeClassifier(face_classifier)
faces = face_cascade.detectMultiScale(cvt, 1.25, 3)
cropped_images=[]
for (x, y, w, h) in faces:
    idx=0
    cv2.rectangle(source, (x, y), (x + w, y + h), (255, 0, 0), 4)
    cropped_img = source[y:y+h, x:x+w]
    cropped_images.append(cropped_img)
    cv2.imwrite('people_face_'+idx+++'.jpg',cropped_img);
    

plt.axis('off')
plt.title('Bouncing Image')
plt.imshow(cv2.cvtColor(source, cv2.COLOR_BGR2RGB))
plt.show()


plt.title('Bounces')
for cropped_img in cropped_images
    plt.imshow(cv2.cvtColor(cropped_img, cv2.COLOR_BGR2RGB))
    plt.show()
```
